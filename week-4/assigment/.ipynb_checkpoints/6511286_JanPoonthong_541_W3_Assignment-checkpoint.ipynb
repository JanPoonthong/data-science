```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("WorldBank_GDP.csv")
df
```

# Ex 1

```python
year_2018_and_2010 = df[(df['Year'] >= 2010) & df["Year"] <= 2018]


grouped_df = year_2018_and_2010.groupby('Country Name')

for name, group in grouped_df:
    plt.plot(group['Year'], group['GDP'], label=name)

plt.legend()
plt.xlabel('Year')
plt.ylabel("GDP")
```

The countried with growing GDP are China and United States, as indicated by the upward trends on the graph
